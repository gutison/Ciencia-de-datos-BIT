```python
"""
Taller 1 - Python Básico
Variables
Crea un programa que permita interactuar con el usuario para recolectar datos. Asegurate que los datos recopilados tengan el formato correcto para poder hacer cálculos.

Solicita al usuario ingresar el nombre de un producto.
Pídele que ingrese el precio del producto y la cantidad vendida.
Convierte el precio a un número decimal (float) y la cantidad a un número entero (int).
Calcula el total de la venta y muestra un mensaje en pantalla con el resumen. 
"""

##invalido
##producto=str
##precio=float
##cantidad=int

producto = input("INGRESE EL NOMBRE DEL PRODUCTO: ")
precio = float(input("INGRESE EL PRECIO DEL PRODUCTO: "))
cantidad = int(input("INGRESE LA CANTIDAD DE PRODUCTOS: "))

total=(precio*cantidad)

print("EL VALOR TOTAL DE LA VENTA ES:", total, "PESOS COLOMBIANOS")





""""
Estructuras de Datos (Listas, Tuplas, Diccionarios)
Crea un programa para organizar información de múltiples ventas utilizando las estructuras adecuadas según la naturaleza de los datos.

Tupla: Crea una tupla con las categorías de productos permitidas en la tienda (ej. "Electrónica", "Ropa", "Alimentos"). Se usa una tupla porque las categorías de la tienda no cambian todos los días.
Diccionario: Crea un diccionario que represente una venta específica. Debe tener las llaves: producto, categoria, precio, cantidad y total.
Lista: Crea una lista vacía llamada registro_ventas y añade el diccionario de la venta que acabas de crear.
"""

categoria = ("Electrónica", "Ropa", "Alimentos")

categoria = input("Ingrese la categoría del producto: ")

venta = {
    "producto":producto,
    "categoria":categoria,
    "precio":precio,
    "cantidad":cantidad,
    "total":total
}
print(venta)

registro_ventas = []

registro_ventas.append(venta)

print(registro_ventas)





"""
Estructuras de Control (Condicionales y Ciclos)
Haz un programa que ayude a automatizar procesos repetitivos y aplicar reglas de negocio (lógica condicional).

Crea un ciclo for que recorra tu lista de registro_ventas (puedes agregar un par de ventas manuales extra al código para hacer la prueba).
Dentro del ciclo, aplica un condicional if: Si el total de la venta es mayor a $100000, aplica un descuento del 10%.
Imprime el reporte final mostrando el producto, el total original y el total con descuento (si aplica).
"""

for venta in registro_ventas:

    if venta["total"] > 100000:

        descuento = venta["total"] * 0.10
        total_con_descuento = venta["total"] - descuento

        print("-------------------------")
        print("Producto:", venta["producto"])
        print("Total original:", venta["total"])
        print("Descuento aplicado: 10%")
        print("Total con descuento:", total_con_descuento)

    else:

        total_con_descuento = venta["total"]

        print("-------------------------")
        print("Producto:", venta["producto"])
        print("Total original:", venta["total"])
        print("No aplica descuento")
        print("Total a pagar:", total_con_descuento)





"""
Funciones y Manejo de Errores
Crea código reutilizable y a prueba de errores de usuario siguiendo estos parámetros.

Crea una función llamada calcular_ingreso que reciba dos parámetros: precio y cantidad.
Dentro de la función, utiliza un bloque try-except. Intenta multiplicar el precio por la cantidad.
Si ocurre un error (por ejemplo, si en lugar de un número llega un texto), el except debe atraparlo, imprimir "Error: Los valores deben ser numéricos" y retornar 0.
Prueba tu función enviando valores correctos y luego enviando un texto para ver cómo se maneja el error.
"""

def calcular_ingreso(precio, cantidad):

    try:
        total = precio * cantidad
        return total

    except:
        print("Error: Los valores deben ser numéricos")
        return 0
```
